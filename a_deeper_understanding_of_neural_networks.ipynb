# A deeper understanding of neurals networks

## Convolutional Neural Networks (CNN)

### Perceptron

The perceptron is a collection of severals units/cells called perceptrons which are **binary linear classifiers**

Inputs **x1** and **x2** are multiplied with theur respective weights **w1** and **w2** and summed togeteher using function **f**, therefore f = x1*w1 + x2*w2 + b, where b is the optional bias term.

This f function is **evaluated** through an activation function which allow for desired classification. 

**Sigmoid** function is the most common activation fucntion used for binary

Now if we stack **multiple inputs** and connect them using the f function with multipole cells stacked in another layer, this forms multiple **fully connected perceptrons** (output on one layer becomes input of the next layer)

Here is an example of the **Simplest Neural Network** == **Feed Foward Neural Net**

Feed foward neural nets are named so since the flow of information is **unidirectional** and **not cyclic**

## Deep Neural Network (Deep Feed Foward Network)

### The CNN Architecture

Cnn are formed using **3** types of layers :
* **Convolution** 
* **Pooling **
* **Dense(Fully Connected)**


An common example of a Dense NNet would be a NNet where all layer are fully connected 

#### Convolution Layer 

Consider an image of 5x5 pixels with **1 as white**  and **0 as black**, the images is recongnized as a monochrome image of dimenssion 5x5. Now imagine a 3X3 matrix with random **1s and 0s** and this matrix is allowed to do a matrix multiplication with a sub-set of image, this multiplication is recorded in new matrix as our 3x3 matrix moves a pixel in every iteration.

The 3x3 matrix moving matrix (yellow) is called a **filter** which has a task to **extract** feature from the image.


It does so by using **Optimization Algorithms**  to decide specific **1s and 0s** in the 3x3 matrix(pink)

A single step of 3X3 filtering is called **Stride**

Here is an example of detailed view 3-channel (RGB) image producing 2 convolution output using two-3channel filter 

The filter **w0 and w1** are the "Convolutions" and the **output is extracted feature**

#### Pooling Layer

This layer is used to **reduce the dimension of input** using different functions.

In general MaxPooling Layer is used after a convolutional layer.

Pooling uses a 2x2 matrix and operates over the image in the same manner as a convolution layer but this time its **reducing the image itself**

#### Dense Layer

This layer is a **fully connected layer** between the activations and the previous layer
