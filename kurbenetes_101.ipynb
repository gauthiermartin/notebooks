# Kubernetes 101

* New standard for deploying and managing software in the cloud


## Hardware

### Nodes


**Node**: Smallest unit of computing in Kubernetes (Normaly a single machine in a cluster)

*  **The nodes communicate with the master using the Kubernetes API**

### Cluster


* The **Master** is responsible for managing the cluster.(Coordinates activites, schedule applications, scalling...)
* Nodes pool together their resources to form a more powerful machine.
* Intelligent work distribution throught nodes on deployement of an application.
* Intelligent load balancing upon adding or removing nodes.

* Kubelets is an agent for managing node and communicating with kuberetes master.

### Presistent Volumes

* Since programs running on a cluster aren't garranteed to run on a specific node , data can't be savec to any arbitrary place in the file system.
* Traditional local storage is theirby treated as a temporary cache.
* Persistent Volumes are used to store data
* They are a file system that could be mounted on the cluster without being associated with any particular node.

## Software

### Containers


* Programs running on kubernetes are packaged as Linux Containers.

* Containerization allows to create self-contained linux execution environments. 
* Containerization implies that application needs to be decoupled from individual host.
* Any program and all its dependencies can be bundeled up into a single file and then shared on the internet.
* Try to limit yourself **1 Process per Container**

## Pods

 * They wrap one or more containers into them
 * All containers in a pod share the same resources and local network
 * Easy communication + isolation from others

* Pods is the unit of replication in kubernetes
* Kubernetes can be configured to deploy new copy of your pod in the cluster if required.
* Allow load balancing / replication
* Pods should remain as small as possible to avoid expensive bills as all container in a pod must be scaled together.

## Deployements

 * Deployment layer is managing the pods
 * Specify how many replica of a pod should be running at the same time
 * When a deployement is added to the cluster == Automatic spin up the requested number of pods
 * If a node goes down or is deleted, kubernetes replace it , as he provides a **self-healing mechanism to address        machine failure or maintenance**.

## Ingress

* Handle external traffic to our application.
* By default pods are isolated from the outside world.
* Channels must be open to enable communications with a pod.


* Default way to add ingress to you cluster 
  1. Ingress controller 
  2. Load Balancer

## Keynotes


 * **Kubernetes automates the distribution and scheduling of application containers across a cluster in a more efficient way**
