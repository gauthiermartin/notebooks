### Drapeau indiquant une fin de trajet
Dans certains cas, on est certain qu'une coordonnée est la dernière d'un trajet:
1. Lorsqu'un prompt est associé: l'usager a indiqué qu'il avait terminé sont trajet
2. Lorsque la coordonnée suivante est hors de la boite d'intérêt (bounding box): à partir de ce contact, on considère le trajet terminé. Une coodonnée est identifiée comme en dehors de la boite d'intérêt lorsque code_rejet==1. On identifie seulement la première coordonnée comme étant la dernièere du trajet.
4. S'il y a plus de 60 minutes entre 2 contacts, on assume une fin de trajet.
5. Le dernier contact d'un usager donné est forcément une fin de trace.


```python
coords['stop'] = np.where(coords.prompt_id >= 0, 1, 0)
coords['stop'] = np.where((coords.code_rejet.isnull())&(coords.code_rejet.shift(-1) == coord_outside_bbox_reject), 1, coords.stop)
```

On impose la règle suivante: s'il y a plus de 60 minutes entre 2 contacts, on assume une fin de trajet.

```python
coords.loc[(coords.delta_time>=max_delta_time),'stop']=1
```

Le dernier contact d'un usager est forcément une fin de trajet:

```python
%%time
last_coords = coords.groupby('uuid', as_index=False).agg({'coordinate_id':'last'}).coordinate_id.values
coords.loc[coords.coordinate_id.isin(last_coords), 'stop'] = 1
```

### Drapeau indiquant si la coordonnée peut être snappée

```python
coords.loc[coords.h_accuracy>poor_accuracy, 'code_rejet'] = coord_poor_precision_reject
```

## Préparation pour l'inférence de fin de trajet et mode
### Définition de la variable de fin de trajet
C'est la variable que l'on veut prédire (y:= end_of_track), celle qui indique si la coordonnée est une fin de trajet ou elle désigne l'état en route.

On a:
* end_of_track == 1 quand la coordonnée est la dernière d'une trace. 
* end_of_track == 0 quand à cette coordonnée l'usager est route. 

On peut tout de suite mettre end_of_track==1 quand stop==1:

```python
coords['end_of_track'] = np.where(coords.stop == 1, 1, np.nan)
```

### Binarisation
On binarise des champs discrets pour les algorithmes d'apprentissage: 

```python
#%%time
#coords = binarize(coords)
```

```python
%%%time
coords['close_to_station'] = dd.from_pandas(coords[['nearest_station','train_type']],npartitions=n_cores).map_partitions(
      lambda df : df.apply(
         lambda x : bin_close_to_station(x),axis=1)).compute(get=get)
```

### Imputation
On impute les Nan/None avec la valeur médiane:

```python
%%time
coords = impute(coords.copy())
coords.sort_values(['uuid','timestamp'], inplace=True)
```
